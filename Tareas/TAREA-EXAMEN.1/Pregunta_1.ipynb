# Pregunta 1

#### Se tiene un oscilador armónico en dos dimensiones (xy) el que la fuerza está dada por $ \vec{F} = −k\vec{r}$ y se tiene una fricción 

#### $$
\vec{f_{\text{fric}}} = \begin{cases}
        −\gamma v \hat{v} \ \ \ \ \ \ \text{ si } \ \ \vert \vec{v} \vert < 1
        \\
        −\mu v^{3/2} \hat{v} \ \ \text{ si } \ \ \vert \vec{v} \vert > 1
        \end{cases}
 $$

#### Para $m = 1$, $\gamma = 0.1$ y $\mu = 0.2$, resuelva los siguientes ejercicios:

#### (a) Encuentre las ecuaciones de movimiento en coordenadas cartesianas y polares.



El oscilador armónico con fricción cumple que:

$$
m\ddot{\vec{r}}-\vec{f}_{\text{fric}}+k\vec{r} = 0  \ \ \ \ \ \ \dots (1)
$$

#### Para cartesianas

Considerando que  $\vec{r}=\sum_{i=1}^2 x_i\hat{e_i} \ $  y escribiendo la fuerza, la fricción y la posición en componentes, se obtiene que:

$$
\ddot{\vec{r}}\hat{e_i}=\ddot{\vec{x_i}}=\ddot{x_i}\hat{e_i}
$$

$$
\vec{F}\hat{e_i}=-k\vec{x_i}=-k x_i\hat{e_i}
$$

$$
\vec{f_{\text{fric}}} \hat{e_i} = \begin{cases}
        −\gamma v \hat{e_i} \ \ \ \ \ \ \text{ si } \ \ \vert \vec{v} \vert < 1
        \\
        −\mu v^{3/2} \hat{e_i} \ \ \text{ si } \ \ \vert \vec{v} \vert > 1
        \end{cases}
$$

Por lo tanto, para $i \in \{ 1,2\}$:

$$
(m\ddot{x_i} -f_{\text{fric}} + k x_i)\hat{e_i} = 0 \ \ \ \ \ \ \ \ \ \ \dots (2)
$$

Definiendo $\beta = \frac{f_{\text{fric}}}{2m} \ \ $ y $ \ \ \omega=\sqrt{\frac{k}{m}} \ \ $, las anteriores ecuaciones se pueden reescribir como:

$$
\ddot{x_i}-2\beta +\omega x_i = 0 
$$

Cada una de éstas ecuaciones diferenciales tiene como solución:

$$
x_i=e^{-\beta t}(C_1 e^{\sqrt{\beta^2-\omega^2}t}+C_2 e^{-\sqrt{\beta^2-\omega^2}t})
$$

Y para resolverlas numéricamente las transformamos en un sistema de ecuaciones, renombrando $x_i=y_i$, $\dot{x_i}=w_i$, entonces:

$$
\dot{y_i}=w_i
$$


$$
\dot{w_i}= \begin{cases}
        −\frac{\gamma \sqrt{w_1^2+w_2^2}}{m} - \omega y_i \ \ \ \ \ \ \text{ si } \ \  \sqrt{w_1^2+w_2^2}  < 1
        \\
        \\
        −\frac{\mu \sqrt{w_1^2+w_2^2}^{3}}{m} - \omega y_i \ \ \text{ si } \ \  \sqrt{w_1^2+w_2^2}  > 1
        \end{cases}
$$


#### Para osos polares


Por otra parte, para coordenadas polares se tiene que:

$$
\vec{r}=r\hat{r}
$$

$$
\dot{\vec{r}}=\dot{r} \hat{r}+r \hat{\theta}
$$

$$
\ddot{\vec{r}} =\ddot{r}\hat{r}+2\dot{r}\hat{\theta}+r\ddot{\hat{r}}
$$

Donde:

$$
\hat{r}=\bigl ( \cos \theta \hat{e_1}+\sin \theta \hat{e_2} \bigr )
$$

$$
\hat{\theta}=\bigl ( -\sin \theta \hat{e_1}+\cos \theta \hat{e_2} \bigr )
$$

$$
\dot{\hat{r}}=\dot{\theta} \bigl ( -\sin \theta \hat{e_1}+\cos \theta \hat{e_2} \bigr ) = \dot{\theta} \hat{ \theta}
$$

$$
\ddot{\hat{r}}=-\dot{\theta}^2\bigl ( \cos \theta \hat{e_1}+\sin \theta \hat{e_2} \bigr )=-\dot{\theta}^2 \hat{r}
$$

Por lo tanto: 

$$
\ddot{\vec{r}} =\ddot{r}\hat{r}+2\dot{r}\hat{\theta}-r\theta^2\hat{r} = (\ddot{r} - r\theta^2 ) \hat{r} + 2\dot{r}\hat{\theta}
$$



Y la fricción queda como:

$$
\vec{f_{\text{fric}}} = \begin{cases}
        −\gamma (\dot{r} \hat{r}+r \hat{\theta}) \ \ \ \ \ \ \text{ si } \ \ \vert r \vert < 1
        \\
        −\mu (\dot{r} \hat{r}+r \hat{\theta})^{3/2} \ \ \text{ si } \ \ \vert r \vert > 1
        \end{cases}
$$

Entonces, la ecuación (1) se convierte en dos ecuaciones dependiendo del valor absoluto de $r$:


$$
\ddot{\vec{r}}-\frac{\vec{f}_{\text{fric}}}{m}+\frac{k}{m}\vec{r} = \begin{cases}
        (\ddot{r} + \frac{\gamma}{m} \dot{r} + (\frac{k}{m} - \dot{\theta}^2)r) \hat{r} + (2\dot{r} + \frac{\gamma}{m} r)\hat{\theta} = 0 \ \ \ \ \ \ \ \text{ si } \ \ \vert r \vert < 1
        \\
        \\
        (\ddot{r} + (\frac{k}{m} - \dot{\theta}^2)r) \hat{r} + 2\dot{r} \dot{\hat{r}} −\frac{\mu}{m} (\dot{r} \hat{r}+r \hat{\theta})^{3/2} = 0 \ \ \ \ \ \text{ si } \ \ \vert r \vert > 1
        \end{cases}
$$

La primera ecuación puede separarse en sus respectivas coordenadas, pero la segunda no se puede separar.

A continuación sólo se muestran los cálculos para la primera ecuación, que se separa en dos ecuaciones, una para la coordenada $\hat{r}$ y otra para la coordenada $\hat{\theta}$:

$$
\ddot{r} + \frac{\gamma}{m} \dot{r} + (\frac{k}{m} - \dot{\theta}^2)r = 0 \ \ \ \ \ \dots(3)
$$

$$
\dot{r} + \frac{\gamma}{m} r = 0 \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \dots(4)
$$

La ecuación (3) con las variables $\beta = \frac{\gamma}{2m}\ $ y $\ \omega=\sqrt{\frac{k - m\dot{\theta}^2}{m}}$ tiene la misma solución que en el caso de las coordenadas cartesianas:

$$
r=e^{-\beta t}(C_1 e^{\sqrt{\beta^2-\omega^2}t}+C_2 e^{-\sqrt{\beta^2-\omega^2}t})
$$

Y la ecuación (4) tiene como solución:

$$
r=C_1e^{-\frac{m}{\gamma}}
$$

Para resolver numéricamente la ec. (2), es necesario convertir la ecuación en un sistema de ecuaciones mediante el siguiente cambio de variables $r=y$, $\dot{r}=w$, entonces:

$$
\dot{y}=w
$$


$$
\dot{w}= - \frac{\gamma}{m} w - (\frac{k}{m} - \dot{\theta}^2)y
$$



#### (b) Utilice el método de Runge-Kutta de 4to orden para encontrar la solución a las ecuaciones de movimiento en el inciso anterior con condiciones iniciales $\vec{r}_0 = (0, 1)$ y $\dot{\vec{r}_0} = (\cos \theta , \sin \theta)$, donde $\theta = n\pi/6$ con $n \in \{0,1,2,3,...,10,11\}$

```python
from pylab import*
%matplotlib inline
```

```python
def rk4(ec,p_ini,x,args=0,h=0.01):
    if (x[1]-x[0]) <= 5.*h:
        h = (x[1]-x[0])/10
    tiempos = arange(x[0],x[1]+h,h)
    sol = zeros((len(tiempos),len(p_ini)))
    sol[0,:] = p_ini
    if args==0:
        for i in xrange(len(tiempos)-1):
            k1 = ec(sol[i,:],tiempos[i])
            k2 = ec(sol[i,:]+0.5*h*k1,tiempos[i]+0.5*h)
            k3 = ec(sol[i,:]+0.5*h*k2,tiempos[i]+0.5*h)
            k4 = ec(sol[i,:]+h*k3,tiempos[i]+h)
            sol[i+1,:]=sol[i,:]+(h/6.)*(k1+2*k2+2*k3+k4)
    else:
        for i in xrange(len(tiempos)-1):
            k1 = ec(sol[i,:],tiempos[i],args)
            k2 = ec(sol[i,:]+0.5*h*k1,tiempos[i]+0.5*h,args)
            k3 = ec(sol[i,:]+0.5*h*k2,tiempos[i]+0.5*h,args)
            k4 = ec(sol[i,:]+h*k3,tiempos[i]+h,args)
            sol[i+1,:]=sol[i,:]+(h/6.)*(k1+2*k2+2*k3+k4)
    return tiempos,sol
```

Primero vamos a definir los dos modelos de fricción:

```python
#v[0]=vel_x, v[1]=vel_y
def fric(v,gamma,mu):
    norma=(v[0]**2+v[1]**2)**0.5
    if norma < 1:
        fric_x=-gamma*norma*v[0]
        fric_y=-gamma*norma*v[1]
        return array([fric_x,fric_y])
    else:
        fric_x=-mu*(norma**(1.5))*v[0]
        fric_y=-mu*(norma**(1.5))*v[1]
        return array([fric_x,fric_y])
```

#### Para cartesianas

Debido a que en coordenadas polares las ecuaciones de movimiento se vuelven muy complicadas, el problema será resuelto únicamente en coordenadas cartesianas. 

A continuación se escriben  las ecuaciones de movimiento descritas en la expresión (2):

```python
#x[0]=posición_x, x[1]=posición_y, x[2]=velocidad_x, x[3]=velocidad_y
#z[0]=gamma, z[1]=mu, z[2]=k, z[3]=m

def ec_mov(x,t,z):
    
    dot_x=x[2]
    fr_x=fric(([x[2],x[3]]),z[0],z[1])
    ddot_x=fr_x[0]/2-(z[2]/z[3])*x[0]
    
    dot_y=x[3]
    fr_y=fric(([x[2],x[3]]),z[0],z[1])
    ddot_y=fr_y[1]/2-(z[2]/z[3])*x[1]
    
    return array([dot_x,dot_y,ddot_x,ddot_y])
```

Para una constante de resorte $k=0.3$, una masa $m=1$ y una velocidad $\dot{\vec{r}_0} = (\cos n\pi/6 , \sin n\pi/6)$ obtenemos, once distintas soluciones (ya que $n \in \{0,1,2,3,...,10,11\}$):

```python
fig = figure(figsize=(14,14))
for j in xrange(12):
    t,sol=rk4(ec_mov,[0,1,cos(j*pi/6),sin(j*pi/6)],[0,100],[0.1,0.2,0.3,1])
    plot(sol[:,0],sol[:,1],label='$v_0=(\cos('+str(j)+'\pi),\sin('+str(j)+'\pi))$')
    legend()
```

#### (c) La energía para las distintas condiciones iniciales y detenga la simulación cuando ésta sea menor al 1% de su valor inicial.

La energía del sistema está dada por la suma de la energía potencial U y la energía cinética K, en éste caso:

$$
U= \frac{1}{2}kr^2=\frac{k}{2}(x_1^2+x_2^2) \ \ \ \ \ \ \ \ \ \ \ \ \ \ K= \frac{1}{2}m\dot{r}^2=\frac{m}{2}(\dot{x_1}^2+\dot{x_2}^2)
$$

Por lo tanto:

$$
E= U+K = \frac{1}{2} (kx_1^2 + m\dot{x_1}^2 + kx_2^2 + m\dot{x_2}^2)
$$

Al resolver las ecuaciones de movimiento con el método runge kutta obtuvimos las posiciones y las velocidades en la direcciones: $\hat{e_1}$ y $\hat{e_2}$, entonces, podemos usar esos resultados para calcular la energía total del sistema.

```python
#x[0]=posicion_x, x[1]=posicion_y, x[2]=vel_x, x[3]=vel_y

#Energia de toda la solución

def E(x,k=0.3,m=1):
    energia=0.5*(k*(x[:,0]**2+x[:,1]**2)+m*(x[:,2]**2+x[:,3]**2))
    return energia
```

```python
#x[0]=posicion_x, x[1]=posicion_y, x[2]=vel_x, x[3]=vel_y

#Energía de solo una de las entradas de la solución

def E1(x,k=0.3,m=1):
    energia=0.5*(k*(x[0]**2+x[1]**2)+m*(x[2]**2+x[3]**2))
    return energia
```

```python
def rk4m(ec,p_ini,x,args=0,h=0.01):
    if (x[1]-x[0]) <= 5.*h:
        h = (x[1]-x[0])/10
    tiempos = arange(x[0],x[1]+h,h)
    sol = zeros((len(tiempos),len(p_ini)))
    sol[0,:] = p_ini
    Ener = zeros((len(tiempos),1))
    E_0=E1(p_ini)
    Ener[0,:]= E_0
    j=0
    for i in xrange(len(tiempos)-1):
        k1 = ec(sol[i,:],tiempos[i],args)
        k2 = ec(sol[i,:]+0.5*h*k1,tiempos[i]+0.5*h,args)
        k3 = ec(sol[i,:]+0.5*h*k2,tiempos[i]+0.5*h,args)
        k4 = ec(sol[i,:]+h*k3,tiempos[i]+h,args)
        sol[j+1,:]=sol[j,:]+(h/6.)*(k1+2*k2+2*k3+k4)
        Ener[j+1,:]=E1(sol[j+1,:])
        if Ener[j+1,:] < E_0*0.01:
            ti =tiempos[:j,]
            so = sol[0:j,:]
            en = Ener[:j,]
            break
        else:
            j+=1
    return ti,so,en
```

#### (d) Dibuje las gráficas del momento angular para los casos anteriores.

El momento angular es:

$$
\vec{L}=\vec{r} \times \vec{p} 
$$

En, nuestro caso $x_3=p_3=0$, entonces:

$$
\vec{L}=L_3=(x_1p_2-x_2p_1)\hat{e_3}
$$

```python
#x[0]=posición_x, x[1]=posición_y, x[2]=vel_x, x[3]=vel_y

def L(x):
    momento = x[:,0]*x[:,3]-x[:,1]*x[:,2]
    return momento
```

```python
fig = figure(figsize=(14,14))
for j in xrange(12):
    t,sol=rk4(ec_mov,[0,1,cos(j*pi/6),sin(j*pi/6)],[0,100],[0.1,0.2,0.3,1])
    plot(L(sol),label='$v_0=(\cos('+str(j)+'\pi),\sin('+str(j)+'\pi))$')
    legend()
```

#### espacio fase $(P_{\rho},\rho)$

En cambio, $P_{\rho}$ corresponde a la derivada de la energía con respecto a $\dot{\rho}$. Si la energía está dada por la expresión:

$$
E= \frac{1}{2}(\dot{x}^2+\dot{y}^2) + \frac{k}{2}(x^2+y^2) = \frac{1}{2}(\dot{\rho}^2+\rho^2\dot{\theta}^2) + \frac{k}{2}\rho^2
$$

Entonces:

$$
P_{\rho} = \frac{\partial E}{\partial \dot{\rho}} = \dot{\rho}
$$

Ahora bien, tenemos que:

$$
\dot{x}+\dot{y} = \dot{\rho} \ \hat{\rho} \ + \ \rho \ \dot{\theta} \ \hat{\theta}
$$

Es decir:

$$
\dot{x}=\dot{\rho}\cos\theta - \rho\dot{\theta} \sin \theta 
$$
$$
\dot{y}=\dot{\rho}\sin\theta + \rho\dot{\theta} \cos \theta
$$

Resolviendo el sistema de ecuaciones obtenemos que:

$$
\dot{\rho}=\frac{\dot{x}x+\dot{y}y}{\rho}=\frac{\dot{x}x+\dot{y}y}{\sqrt{x^2+y^2}}
$$

Entonces:

$$
P_{\rho} = \frac{\dot{x}x+\dot{y}y}{\sqrt{x^2+y^2}}
$$

```python
def rho(x,y):
    return (x**2+y**2)**(0.5)
def theta(x,y):
    return arctan(y/x)
def p_rho(x,y,v_x,v_y):
    return (v_x*x+v_y*y)/rho(x,y)
```
