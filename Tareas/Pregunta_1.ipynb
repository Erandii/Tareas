#### Se tiene un oscilador armónico en dos dimensiones (xy) el que la fuerza está dada por $ \vec{F} = −k\vec{r}$ y se tiene una fricción 

 $$
\vec{f_{\text{fric}}} = \begin{cases}
        −\gamma v \hat{v} \ \ \ \ \ \ \text{ si } \ \ \vert \vec{v} \vert < 1
        \\
        −\mu v^{3/2} \hat{v} \ \ \text{ si } \ \ \vert \vec{v} \vert > 1
        \end{cases}
 $$

#### Para $m = 1$, $\gamma = 0.1$ y $\mu = 0.2$, resuelva los siguientes ejercicios:

### (a) Encuentre las ecuaciones de movimiento en coordenadas cartesianas y polares.



El oscilador armónico con fricción cumple que:

$$
m\ddot{\vec{r}}-\vec{f}_{\text{fric}}+k\vec{r} = 0  \ \ \ \ \ \ \dots (1)
$$

#### Para cartesianas

En el caso de coordenadas cartesianas, reescribiendo a la fuerza, la fricción y la posición en componentes:

Sea $\vec{r}=\sum_{i=1}^2 x_i\hat{e_i}$, entonces:


$$
\ddot{\vec{r_i}}=\ddot{\vec{x_i}}=\ddot{x_i}\hat{e_i}
$$

$$
\vec{F_i}=-k\vec{x_i}=-k x_i\hat{e_i}
$$

$$
\vec{f_{\text{fric}}}_i = \begin{cases}
        −\gamma \dot{x_i} \hat{e_i} \ \ \ \ \ \ \text{ si } \ \ \vert \vec{x_i} \vert < 1
        \\
        −\mu \dot{x_i}^{3/2} \hat{e_i} \ \ \text{ si } \ \ \vert \vec{x_i} \vert > 1
        \end{cases}
$$

Definiendo $\beta_i = \frac{f_{{\text{fric}}_i}}{2m} \ \ $ y $ \ \ \omega=\sqrt{\frac{k}{m}} \ \ $, la ecuación (1) se puede reescribir como:

$$
\ddot{x_i}-2\beta_i +\omega x_i = 0 
$$

Cada una de éstas ecuaciones diferenciales tiene como solución:

$$
x_i=e^{-\beta t}(C_1 e^{\sqrt{\beta^2-\omega^2}t}+C_2 e^{-\sqrt{\beta^2-\omega^2}t})
$$

Y para resolverlas numéricamente las transformamos en un sistema de ecuaciones, renombrando $x_i=y_i$, $\dot{x_i}=w_i$, entonces:

$$
\dot{y_i}=w_i
$$


$$
\dot{w_i}= \begin{cases}
        −\gamma w_i / m - \omega y_i \ \ \ \ \ \ \text{ si } \ \ \vert \vec{x_i} \vert < 1
        \\
        −\mu \dot{w_i}^{3/2} / m - \omega y_i \ \ \text{ si } \ \ \vert \vec{x_i} \vert > 1
        \end{cases}
$$


#### Para osos polares


Por otra parte, para coordenadas polares se tiene que:

$$
\vec{r}=r\hat{r}
$$

$$
\dot{\vec{r}}=\dot{r} \hat{r}+r \hat{\theta}
$$

$$
\ddot{\vec{r}} =\ddot{r}\hat{r}+2\dot{r}\hat{\theta}+r\ddot{\hat{r}}
$$

Donde:

$$
\hat{r}=\bigl ( \cos \theta \hat{e_1}+\sin \theta \hat{e_2} \bigr )
$$

$$
\hat{\theta}=\bigl ( -\sin \theta \hat{e_1}+\cos \theta \hat{e_2} \bigr )
$$

$$
\dot{\hat{r}}=\dot{\theta} \bigl ( -\sin \theta \hat{e_1}+\cos \theta \hat{e_2} \bigr ) = \dot{\theta} \hat{ \theta}
$$

$$
\ddot{\hat{r}}=-\dot{\theta}^2\bigl ( \cos \theta \hat{e_1}+\sin \theta \hat{e_2} \bigr )=-\dot{\theta}^2 \hat{r}
$$

Por lo tanto: 

$$
\ddot{\vec{r}} =\ddot{r}\hat{r}+2\dot{r}\hat{\theta}-r\theta^2\hat{r} = (\ddot{r} - r\theta^2 ) \hat{r} + 2\dot{r}\hat{\theta}
$$



Y la fricción queda como:

$$
\vec{f_{\text{fric}}} = \begin{cases}
        −\gamma (\dot{r} \hat{r}+r \hat{\theta}) \ \ \ \ \ \ \text{ si } \ \ \vert r \vert < 1
        \\
        −\mu (\dot{r} \hat{r}+r \hat{\theta})^{3/2} \ \ \text{ si } \ \ \vert r \vert > 1
        \end{cases}
$$

Por lo tanto, la ecuación (1) se convierte en dos ecuaciones dependiendo del valor absoluto de $r$:


$$
\ddot{\vec{r}}-\frac{\vec{f}_{\text{fric}}}{m}+\frac{k}{m}\vec{r} = \begin{cases}
        (\ddot{r} + \frac{\gamma}{m} \dot{r} + (\frac{k}{m} - \dot{\theta}^2)r) \hat{r} + (2\dot{r} + \frac{\gamma}{m} r)\hat{\theta} = 0 \ \ \ \ \ \ \ \text{ si } \ \ \vert r \vert < 1
        \\
        \\
        (\ddot{r} + (\frac{k}{m} - \dot{\theta}^2)r) \hat{r} + 2\dot{r} \dot{\hat{r}} −\frac{\mu}{m} (\dot{r} \hat{r}+r \hat{\theta})^{3/2} = 0 \ \ \ \ \ \text{ si } \ \ \vert r \vert > 1
        \end{cases}
$$

La primera ecuación puede separarse en sus respectivas coordenadas, quedando como:

$$
\ddot{r} + \frac{\gamma}{m} \dot{r} + (\frac{k}{m} - \dot{\theta}^2)r = 0 \ \ \ \ \ \dots(2)
$$

$$
\dot{r} + \frac{\gamma}{m} r = 0 \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \dots(3)
$$

La ecuación (2) con las variables $\beta = \frac{\gamma}{2m}\ $ y $\ \omega=\sqrt{\frac{k - m\dot{\theta}^2}{m}}$ tiene la misma solución que en el caso de las coordenadas cartesianas:

$$
r=e^{-\beta t}(C_1 e^{\sqrt{\beta^2-\omega^2}t}+C_2 e^{-\sqrt{\beta^2-\omega^2}t})
$$

Y la ecuación (3) tiene como solución:

$$
r=C_1e^{-\frac{m}{\gamma}}
$$

Para resolver numéricamente la ec. (2), es necesario convertir la ecuación en un sistema de  $r=y$, $\dot{r}=w$, entonces:

$$
\dot{y}=w
$$


$$
\dot{w}= - \frac{\gamma}{m} w - (\frac{k}{m} - \dot{\theta}^2)y
$$



### (b) Utilice el método de Runge-Kutta de 4to orden para encontrar la solución a las ecuaciones de movimiento en el inciso anterior con condiciones iniciales $\vec{r}_0 = (0, 1)$ y $\vec{r}_0 = (\cos \theta , \sin \theta)$, donde $\theta = n\pi$ con $n \in \{0,1,2,3,...,10,11\}$

```python
from pylab import*
%matplotlib inline
```

```python
def rk4(ec,p_ini,x,args=0,h=0.01):
    if (x[1]-x[0]) <= 5.*h:
        h = (x[1]-x[0])/10
    tiempos = arange(x[0],x[1]+h,h)
    sol = zeros((len(tiempos),len(p_ini)))
    sol[0,:] = p_ini
    if args==0:
        for i in xrange(len(tiempos)-1):
            k1 = ec(sol[i,:],tiempos[i])
            k2 = ec(sol[i,:]+0.5*h*k1,tiempos[i]+0.5*h)
            k3 = ec(sol[i,:]+0.5*h*k2,tiempos[i]+0.5*h)
            k4 = ec(sol[i,:]+h*k3,tiempos[i]+h)
            sol[i+1,:]=sol[i,:]+(h/6.)*(k1+2*k2+2*k3+k4)
    else:
        for i in xrange(len(tiempos)-1):
            k1 = ec(sol[i,:],tiempos[i],args)
            k2 = ec(sol[i,:]+0.5*h*k1,tiempos[i]+0.5*h,args)
            k3 = ec(sol[i,:]+0.5*h*k2,tiempos[i]+0.5*h,args)
            k4 = ec(sol[i,:]+h*k3,tiempos[i]+h,args)
            sol[i+1,:]=sol[i,:]+(h/6.)*(k1+2*k2+2*k3+k4)
    return tiempos,sol
```

```python
def fric(gamma,mu,v):
    if abs(v) < 1:
        return -gamma*v
    else:
        return -mu*v**(1.5)
```

```python
def omega(k,m):
    return (k/m)**0.5
```

#### Para cartesianas

```python
#x[0]=posicion, x[1]=velocidad
#z[0]=gamma, z[1]=mu, z[2]=k, z[3]=m
def ec_car(x,t,z):
    return array([x[1],fric(z[0],z[1],x[1])/2-omega(z[2],z[3])*x[0]])
```

```python
t,sol1=rk4(ec_car,[0,1],[0,100],[0.1,0.2,1,1])
plot(t,sol1)
show()
```

```python
t,sol2=rk4(ec_car,[1,1],[0,100],[0.1,0.2,1,1])
plot(t,sol2)
show()
```

```python
plot(sol1,sol2)
show()
```

#### Para polares

```python
#z[0]=gamma, z[1]=mu, z[2]=k, z[3]=m
def ec_pol1(y,t,z):
    return array([y[1],fric(z[0],z[1],y[1])/z[3]-z[2]*y[0]/z[3]])
def ec_pol2(x,t,z):
    return fric(z[0],z[1],x[1])/z[3]
```

```python
t,res1=rk4(ec_pol1,[cos(8*pi),-sin(8*pi)],[0,100],[0.1,0.2,0.5,1])
t,res2=rk4(ec_pol2,[sin(4*pi),cos(4*pi)],[0,100],[0.1,0.2,0.5,1])
plot(t,res1)
show()
```

```python
plot(t,res2)
show()
```
